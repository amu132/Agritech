```python
import pandas as pd
import numpy as np
import json
import sys
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Read the dataset
dataset = pd.read_csv('/content/Crop_recommendation (1).csv')

# Display first 5 rows to understand the dataset structure
dataset.head()

# Get command-line arguments or set default values if not provided
if len(sys.argv) >= 8:  # Check if enough arguments are provided
    jsonn = sys.argv[1]
    jsonp = sys.argv[2]
    jsonk = sys.argv[3]
    jsont = sys.argv[4]
    jsonh = sys.argv[5]
    jsonph = sys.argv[6]
    jsonr = sys.argv[7]
else:
    # Set default values if arguments are not provided
    print("Not enough command-line arguments provided. Using default values.")
    jsonn = 1  # Replace with your desired default value
    jsonp = 2  # Replace with your desired default value
    jsonk = 3  # Replace with your desired default value
    jsont = 4  # Replace with your desired default value
    jsonh = 5  # Replace with your desired default value
    jsonph = 6  # Replace with your desired default value
    jsonr = 7  # Replace with your desired default value
```

```python
# Divide dataset into features (X) and labels (y)
X = dataset.iloc[:, :-1].values  # All columns except last (features)
y = dataset.iloc[:, -1].values   # Last column (crop label)

# Split dataset into training (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initialize and train the Random Forest Classifier
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
```

```python
# Make predictions on test data
y_pred = classifier.predict(X_test)

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the dataset
dataset = pd.read_csv('/content/Crop_recommendation (1).csv')

# Split dataset into features (X) and labels (y)
X = dataset.iloc[:, :-1].values  # All columns except last
y = dataset.iloc[:, -1].values   # Last column (crop label)

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the Random Forest model
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)

# Evaluate model performance
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Define user input values manually (Replace with actual values)
n_params = 90    # Example: Nitrogen
p_params = 42    # Example: Phosphorus
k_params = 43    # Example: Potassium
t_params = 21    # Example: Temperature
h_params = 82    # Example: Humidity
ph_params = 6.5  # Example: pH
r_params = 203   # Example: Rainfall

# Convert values into numpy array for prediction
user_input = np.array([[n_params, p_params, k_params, t_params, h_params, ph_params, r_params]])

# Make prediction
predictions = classifier.predict(user_input)
print(f"\nRecommended Crop: {predictions[0]}")
```
